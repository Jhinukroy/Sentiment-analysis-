```python
from transformers import pipeline

# Load the sentiment analysis model
sentiment_analyser = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment")

# Define a mapping from model labels to human-readable sentiment
label_mapping = {
    'LABEL_0': 'negative',
    'LABEL_1': 'neutral',
    'LABEL_2': 'positive'
}
```

```python
# Prompt the user for input
user_input = input("Please enter a sentence to analyze sentiment: ")

# Perform sentiment analysis on the user's input
result = sentiment_analyser(user_input)

# Extract the label and score
label = result[0]['label']
score = result[0]['score']

# Convert the label to human-readable text
sentiment = label_mapping.get(label, 'unknown')


# Output the result
print(f"Sentiment analysis result: {sentiment} (Score: {score:.2f})")
```
